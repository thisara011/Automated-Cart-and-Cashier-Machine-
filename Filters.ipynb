```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image and convert to grayscale
text1 = cv2.imread('/content/sample_data/Screenshot 2024-06-07 140021.png', cv2.IMREAD_GRAYSCALE)

# Define and normalize the 5x5 kernel
kernel_5x5 = np.ones((5, 5), np.float32) / (5 * 5)

# Define and normalize the 23x23 kernel
kernel_23x23 = np.ones((23, 23), np.float32) / (23 * 23)

# Apply the 5x5 filter
filter_img = cv2.filter2D(text1, -1, kernel_5x5)

# Apply the 23x23 filter
filter_img02 = cv2.filter2D(text1, -1, kernel_23x23)

# Display the original and filtered images side by side
cv2_imshow(np.hstack((text1, filter_img, filter_img02)))
```
